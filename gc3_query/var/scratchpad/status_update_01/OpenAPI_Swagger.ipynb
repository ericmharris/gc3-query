# Swagger/OpenAPI

[https://swagger.io/docs/specification/about/](https://swagger.io/docs/specification/about/)

<h3>What Is OpenAPI?</h3>
<p><b>OpenAPI Specification</b> (formerly Swagger Specification) is an API description format for REST APIs. An OpenAPI file allows you to describe your entire API, including:</p>
<ul>
<li>Available endpoints (<code>/users</code>) and operations on each endpoint (<code>GET /users</code>, <code>POST /users</code>)</li>
<li>Operation parameters Input and output for each operation</li>
<li>Authentication methods</li>
<li>Contact information, license, terms of use and other information.</li>
</ul>
<p>API specifications can be written in YAML or JSON. The format is easy to learn and readable to both humans and machines.</p>
<p>The complete OpenAPI Specification can be found on GitHub:<br>
<a href="https://github.com/OAI/OpenAPI-Specification/blob/master/versions/3.0.0.md">OpenAPI 3.0 Specification</a>

















<h2 id="basic-structure-0" class="active">Basic Structure of Swagger Spec</h2>


[https://swagger.io/docs/specification/basic-structure/](https://swagger.io/docs/specification/basic-structure/)


<p>You can write OpenAPI definitions in <a href="https://en.wikipedia.org/wiki/YAML">YAML</a> or <a href="https://en.wikipedia.org/wiki/JSON">JSON</a>. In this guide, we use only YAML examples but JSON works equally well.</p>
<p>A sample OpenAPI 3.0 definition written in YAML looks like:</p>
<pre><code>openapi: 3.0.0
<strong>info</strong>:
  title: Sample API
  description: Optional multiline or single-line description in [CommonMark](http://commonmark.org/help/) or HTML.
  version: 0.1.9

servers:
  - url: http://api.example.com/v1
    description: Optional server description, e.g. Main (production) server
  - url: http://staging-api.example.com
    description: Optional server description, e.g. Internal staging server for testing

<strong>paths</strong>:
  /users:
<strong>    get</strong>:
      summary: Returns a list of users.
      description: Optional extended description in CommonMark or HTML.
<strong>      responses</strong>:
        '200':    # status code
          description: A JSON array of user names
          content:
            application/json:
              schema: 
                type: array
                items: 
                  type: string
</code></pre>
<p><!--As you can see, the specification has a <code>keyword: value</code> format. We often call the keyword part <i>field</i>, <i>property</i> or <i>node</i>. Keywords that have lots of child items are also called <i>sections</i>.--></p>



<h3 id="metadata-1" class="active">Metadata</h3>
<p>Every OpenAPI specification starts with the <code>openapi</code> keyword mentioning the version of the specification format. This version defines the overall structure of an API specification – what you can document and how you document it.</p>
<p>Earlier, the format field included two components (for example, 2.0). Since version 3, OpenAPI is using <a href="http://semver.org/">semantic versioning</a> with three components. The latest version is <code>3.0.0</code>:</p>
<pre><code>openapi: 3.0.0</code></pre>
<p>The <code>info</code> section contains API information: <code>title</code>, <code>description</code> (optional), <code>version</code>:</p>
<pre><code>info:
  title: Sample API
  description: Optional multiline or single-line description in [CommonMark](http://commonmark.org/help/) or HTML.
  version: 0.1.9
</code></pre>
<p><code>title</code> is your API name.</p>
<p><code>description</code> is extended information about your API. It can be <a href="http://stackoverflow.com/a/21699210">multiline</a> and supports the <a href="http://commonmark.org/help/">CommonMark</a> dialect of Markdown for rich text representation. HTML is supported to the extent provided by CommonMark (see <a href="http://spec.commonmark.org/0.27/">HTML Blocks</a><a> in </a><a href="http://spec.commonmark.org/0.27/">CommonMark 0.27 Specification</a><a>).</a></p>
<p><code>version</code> is an arbitrary string that specifies the version of your API (do not confuse it with file revision or the <code>openapi</code> version). You can use <a href="http://semver.org/">semantic versioning</a> like <em>major.minor.patch</em>, or an arbitrary string like <em>1.0-beta</em> or <em>2017-07-25</em>.</p>
<p><code>info</code> also supports other keywords for contact information, license, terms of service, and other details. <!--See <a href="/docs/specification/metadata">Metadata</a> for details.--></p>



<h3 id="servers-2" class="active">Servers</h3>
<p>The <code>servers</code> section specifies the API server and base URL. You can define one or several servers, such as production and sandbox.</p>
<pre><code>servers:
  - url: http://api.example.com/v1
    description: Optional server description, e.g. Main (production) server
  - url: http://staging-api.example.com
    description: Optional server description, e.g. Internal staging server for testing
</code></pre>
<p>All API paths are relative to the server URL. In the example above, <code>/users</code> means <code>http://api.example.com/v1/users</code> or <code>http://staging-api.example.com/v1/users</code>, depending on the server used.<br>

<p>For more information, see <a href="/docs/specification/api-host-and-base-path/">API Server and Base Path</a>.</p>

<h3 id="paths-3" class="active">Paths</h3>
<p>The <code>paths</code> section defines individual endpoints (paths) in your API, and the HTTP methods (operations) supported by these endpoints. For example, <code>GET /users</code> can be described as:</p>
<pre><code>paths:
  /users:
    get:
      summary: Returns a list of users.
      description: Optional extended description in CommonMark or HTML
      responses:
        '200':
          description: A JSON array of user names
          content:
            application/json:
              schema: 
                type: array
                items: 
                  type: string
</code></pre>
<p>An operation definition includes parameters, request body (if any), possible response status codes (such as 200 OK or 404 Not Found) and response contents.<br>

For more information, see <a href="/docs/specification/paths-and-operations/">Paths and Operations</a>.</p>

<h3 id="parameters-4" class="active">Parameters</h3>
<p>Operations can have parameters passed via URL path (<code>/users/{userId}</code>), query string (<code>/users?role=admin</code>), headers (<code>X-CustomHeader: Value</code>) or cookies (<code>Cookie: debug=0</code>). You can define the parameter data types, format, whether they are required or optional, and other details:</p>
<pre><code>paths:
  /user/{userId}:
    get:
      summary: Returns a user by ID.
      parameters:
        - name: userId
          in: path
          required: true
          description: Parameter description in CommonMark or HTML.
          schema:
            type : integer
            format: int64
            minimum: 1
      responses: 
        '200':
          description: OK
</code></pre>
<p>For more information, see <a href="/docs/specification/describing-parameters/">Describing Paramet

<h3 id="request-body-5" class="active">Request Body</h3>
<p>If an operation sends a request body, use the <code>requestBody</code> keyword to describe the body content and media type.</p>
<pre><code>paths:
  /users:
    post:
      summary: Creates a user.
      requestBody:
        required: true
        content:
          application/json:
            schema:
              type: object
              properties:
                username:
                  type: string
      responses: 
        '201':
          description: Created
</code></pre>
<p>For more information, see <a href="/docs/specification/describing-request-body/">Describing Request Body</a>.</p>

<h3 id="responses-6" class="active">Responses</h3>
<p>For each operation, you can define possible status codes, such as 200 OK or 404 Not Found, and the response body <code>schema</code>. Schemas can be defined inline or referenced via <a href="/docs/specification/using-ref/"><code>$ref</code></a>. <!-- (see below).--> You can also provide example responses for different content types:</p>
<pre><code>paths:
  /user/{userId}:
    get:
      summary: Returns a user by ID.
      parameters:
        - name: userId
          in: path
          required: true
          description: The ID of the user to return.
          schema:
            type: integer
            format: int64
            minimum: 1
      responses:
        '200':
          description: A user object.
          content:
            application/json:
              schema:
                type: object
                properties:
                  id:
                    type: integer
                    format: int64
                    example: 4
                  name:
                    type: string
                    example: Jessica Smith
        '400':
          description: The specified user ID is invalid (not a number).
        '404':
          description: A user with the specified ID was not found.
        default:
          description: Unexpected error
</code></pre>
<p>Note that the response HTTP status codes must be enclosed in quotes: “200” (OpenAPI 2.0 did not require this).<br>

# Bravado Swagger Client

- Maintained by Yelp
- Highly configurable, very well tested, very large userbase
- Plugable back-end for HTTP client libraries





<a class="headerlink" href="#bravado-documentation" title="Permalink to this headline">¶</a></h1>
<p><strong>Bravado</strong> is a python client library for Swagger 2.0 services.</p>
<p>More information on Swagger can be found <a class="reference external" href="http://swagger.io/">on the Swagger website</a></p>
<p>It aims to be a complete replacement to <a class="reference external" href="https://github.com/wordnik/swagger-codegen">swagger codegen</a>.</p>
<p>Features include:</p>
<ul class="simple">
<li>Dynamically generated client - no code generation needed!</li>
<li><a class="reference external" href="http://docs.python-requests.org/en/latest/">Synchronous</a> and <a class="reference external" href="https://github.com/Yelp/fido">Asynchronous</a> http clients out of the box.</li>
<li>Strict validations to verify that your Swagger Schema is  <a class="reference external" href="https://github.com/wordnik/swagger-spec/blob/master/versions/2.0.md/">v2.0</a> compatible.</li>
<li>HTTP request and response validation against your Swagger Schema.</li>
<li>Swagger models as Python types (no need to deal with JSON).</li>
<li>REPL friendly navigation of your Swagger schema with docstrings for Resources, Operations and Models.</li>
<li>Ingestion of your Swagger schema via http or a local file path.</li>
