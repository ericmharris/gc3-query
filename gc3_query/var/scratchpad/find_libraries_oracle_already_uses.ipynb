```python
# Split jinja_var into a list of valid Python variable/function names (alphanumeric + _)
IMPORT_RE = re.compile(r"""
                                ^                             # line start
                                \s*                             # zero or more spaces
                                import                             # from
                                \s+                             # one or more spaces
                               (?P<library_name>[_A-Za-z]         # first char is alpha or _
                               [_a-zA-Z0-9]*)                   # zero or more alpha-num or _
                               \s*                              # zero or more spaces
                            """, re.VERBOSE)
```

```python
MULTI_IMPORT_RE = re.compile(r"""
                                ^                             # line start
                                \s*                             # zero or more spaces
                                import                             # from
                                \s+                             # one or more spaces
                               (?P<library_name>[_A-Za-z]         # first char is alpha or _
                               [_a-zA-Z0-9]*)                   # zero or more alpha-num or _
                               \s*                              # zero or more spaces
                               ,                              # comma
                            """, re.VERBOSE)
```

```python
s = r'import json, os, sys'
if MULTI_IMPORT_RE.match(s):
    import_part, *other_libraries = s.split(',')
    other_libraries = [s.strip() for s in other_libraries]
    print(f'import_part={import_part}, other_libraries={other_libraries}')

    
    
def parse_import_libraries(line: str) -> List[str]:
    """find libraries in 'import os, sys, json' or 'import os' """
    if MULTI_IMPORT_RE.match(line):
        import_part, *other_libraries = s.split(',')
        other_libraries = [s.strip() for s in other_libraries]
        other_libraries.append(import_part.split()[1])
        imported_libraries = [l for l in other_libraries if is_library_name_valid(l)]
        return imported_libraries
    if IMPORT_RE.match(line):
        imported_libraries = [l for l in IMPORT_RE.findall(line) if is_library_name_valid(l)]
        return imported_libraries
    if FROM_IMPORT_RE.match(line):
        imported_libraries = [l for l in FROM_IMPORT_RE.findall(line) if is_library_name_valid(l)]
        return imported_libraries


```

```python
all_imported_libs: Dict[Path, List[ImportedLibrary]] = {}

# for py_file in BASE_DIR.joinpath('psmcli').rglob('*.py'):
for py_file in BASE_DIR.rglob('*.py'):
    lib_names_in_file = []
    try:
        lines = py_file.read_text().splitlines()
    except UnicodeDecodeError as e:
        print(f"UnicodeDecodeError reading {py_file}:\n{e}")
        
    for line in lines:
        if is_import_line(line):
            try:
                lib_names = parse_import_libraries(line)
                lib_names_in_file.extend(lib_names)
            except TypeError as e:
                print(f"TypeError importing {py_file}:\n{e}")
                
                
    all_imported_libs[py_file] = [ImportedLibrary(file_path=py_file, library_name=name) for name in lib_names_in_file]
    print(f"Found {len(lib_names_in_file)}/{len(all_imported_libs)} imports in {py_file}")

            
```

```python
unique_libraries = set()
library_found_in: Dict[str, WindowsPath] = {}

for py_file, imported_libraries in all_imported_libs.items():
    lib_names_in_py_file = [il.library_name for il in imported_libraries]
    for lib_name in lib_names_in_py_file:
        if lib_name not in unique_libraries:
#             print(f"Adding {lib_name} ")
            unique_libraries.add(lib_name)
            library_found_in[lib_name] = [py_file]
        else:
            library_found_in[lib_name].append([py_file])

print(f"Found {len(unique_libraries)} in {len(all_imported_libs)}")
unique_libraries_lower_case = set([name.lower() for name in unique_libraries])
```

```python
for lib_name in sorted(list(unique_libraries)):
    f = dict(lib_name=lib_name,  library_found_in=library_found_in)
    print("{:40} found in {:3} files: {}".format(lib_name, len(library_found_in[lib_name]),library_found_in[lib_name][0] ))
```
